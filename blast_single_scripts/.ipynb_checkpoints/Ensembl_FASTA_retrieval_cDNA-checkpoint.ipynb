# FASTA download from Ensembl Server

### Download the FASTA files

The server provides the same structure to find the files. Maybe we can automate the retrieval where:
1. the first script scrapes the names<br>
    Find server address
    ftp://ftp.ensemblgenomes.org/pub/release-40/plants/fasta/arabidopsis_thaliana/cdna/<br>
    The variables in this are:<br>
    RELEASE: release-40 (new releases will have higher numbers e.g. current release = 44)<br>
    DATA: cdna or pep <br>
    SPECIES: appear at two points in the address and can be optained from the `taxa` table `species_name` <br>
    Then scrape the name of the file from the server page that ends with *.cdna.all.fa.gz or *.pep.all.fa.gz respecively


2. the second script generatea the curl query (should look like below)



3. the third script executes them

## Build the Databases

The cDNA files are then collected and converted to a species database that can be BLASTED (see blast script)

This can be done automatically similar to the above, where the `taxa` table is used to generate the query using the cdna files and filling in the taxa numbers into the BLASTdb names: blastdb_3704_nucl as well as the type of data:

cdna: nucl in the name so we know which database it is and -dbtype = nucl so the computer knows which one it is.<br>
pep: prot in the name so we know which database it is and -dbtype = prot so the computer knows which one it is.

!gzcat Arabidopsis_thaliana.TAIR10.cdna.all.fa.gz |makeblastdb -in - -out blastdb_3704_nucl -input_type=fasta -dbtype nucl -title athaliana

-title: can extract the `species_name` from `taxa`. This is only informative and not rigit
